# tiktoken - BPE Tokenizer of GPT-2/3

```python
import tiktoken # pip install tiktoken
```

```python
text1 = "Hello, how are you?"
text2 = "I am Ahammed Shaneeb."
text3 = "Yourfullname please"

all_text = " <|endoftext|> ".join([text1, text2, text3])

print(all_text)
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
token_ids = tokenizer.encode(all_text, allowed_special={"<|endoftext|>"})

print(f"Token IDs:\n{token_ids}")
```

* No error is obtained in relation to OOV (Out of Vocabulary)
* endoftext token's id is at the/almost end id in the vocab
* yourfullname wouldn't be treated as single token

```python
decoded_text = tokenizer.decode(token_ids)
print(f"Decoded text from all text token IDs:\n{decoded_text}")
```
